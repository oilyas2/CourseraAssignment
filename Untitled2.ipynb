```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.feature_selection import RFE

# Load dataset
merged_df = pd.read_csv('merged_data.csv')

# Drop irrelevant columns and handle missing values
merged_df.drop(columns=["rain"], inplace=True)
merged_df.dropna(subset=["price"], inplace=True)

# Encode categorical columns
label_enc = LabelEncoder()
categorical_columns = ["cab_type","source", "destination", "name", "location", "product_id", "id"]  # Removed "cab_type"
for col in categorical_columns:
    merged_df[col] = label_enc.fit_transform(merged_df[col])

# Identify numerical features
numerical_cols = ["distance", "surge_multiplier"]

# Normalize all numerical features in selected_features
scaler = StandardScaler()
selected_numerical_features = [feat for feat in numerical_cols if feat in merged_df.columns]
merged_df[selected_numerical_features] = scaler.fit_transform(merged_df[selected_numerical_features])

# Remove Outliers using IQR Filtering
Q1 = merged_df["price"].quantile(0.25)
Q3 = merged_df["price"].quantile(0.75)
IQR = Q3 - Q1
merged_df = merged_df[merged_df["price"] < (Q3 + 1.5 * IQR)]
print("New dataset size after removing outliers:", merged_df.shape)

# Define model for feature selection
model = Ridge(alpha=5.0)  # Increased Ridge Regularization to reduce overfitting

# Set up RFE to select the top 5 features
rfe = RFE(model, n_features_to_select=5)

# Select features
X = merged_df.drop(columns=["price"])  # Exclude price (target)
y = merged_df["price"]
X_rfe = rfe.fit_transform(X, y)

# Extract selected feature names
selected_features = X.columns[rfe.support_]
print("Selected features using RFE:", list(selected_features))

# Normalize only numerical selected features
selected_numerical_features = [feat for feat in selected_features if feat in numerical_cols]
merged_df[selected_numerical_features] = scaler.fit_transform(merged_df[selected_numerical_features])

# Step 1: Use K-Means to Cluster Rides (k=6 is optimal)
kmeans = KMeans(n_clusters=6, random_state=42)
merged_df["cluster"] = kmeans.fit_predict(merged_df[selected_features])

# Step 2: Train a Separate Ridge Regression Model for Each Cluster with Train-Test-CV Split
models = {}
mae_scores = {}

for cluster in merged_df["cluster"].unique():
    cluster_data = merged_df[merged_df["cluster"] == cluster]
    X_cluster = cluster_data[selected_features]  # Use only selected features
    y_cluster = cluster_data["price"]

    # Train-test split (70% train, 15% test, 15% cross-validation)
    X_train, X_temp, y_train, y_temp = train_test_split(X_cluster, y_cluster, test_size=0.30, random_state=42)
    X_test, X_cv, y_test, y_cv = train_test_split(X_temp, y_temp, test_size=0.50, random_state=42)

    model = Ridge(alpha=5.0)

    # Cross-validation (15% of data)
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    cv_mae = -np.mean(cross_val_score(model, X_cv, y_cv, cv=kf, scoring="neg_mean_absolute_error"))

    # Train model on 70% training data
    model.fit(X_train, y_train)

    # Predict on 15% test data and evaluate
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)

    # Store model and performance
    models[cluster] = model
    mae_scores[cluster] = {"test_MAE": mae, "cv_MAE": cv_mae}

# Output MAE scores for each cluster
print("Mean Absolute Error per cluster (Test & CV):", mae_scores)

# Step 3: Analyze Feature Importance Per Cluster
for cluster in models.keys():
    print(f"Feature Importance for Cluster {cluster}:")
    print(pd.DataFrame({"Feature": selected_features, "Coefficient": models[cluster].coef_})
          .sort_values(by="Coefficient", ascending=False))

# Step 4: Plot Residuals Per Cluster
for cluster in models.keys():
    y_pred = models[cluster].predict(merged_df[selected_features])
    residuals = merged_df["price"] - y_pred

    plt.figure(figsize=(6, 4))
    plt.hist(residuals, bins=30, edgecolor="black")
    plt.title(f"Residuals for Cluster {cluster}")
    plt.xlabel("Prediction Error")
    plt.ylabel("Frequency")
    plt.show()
```
